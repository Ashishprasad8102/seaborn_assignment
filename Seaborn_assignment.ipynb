```python
1> Bar Plot:

Use: Bar plots are used to display categorical data, where one axis represents the categories and the other
axis represents a numerical value. Each bar's height corresponds to the value of the category it represents.
Common Use Cases: Comparing data across different categories, displaying frequency counts, visualizing survey
results, etc.


Histogram:

Use: Histograms are used to visualize the distribution of a continuous numerical variable. They represent
data by dividing it into intervals (bins) along the x-axis and showing the frequency of data points within
each bin on the y-axis.

Common Use Cases: Analyzing data distribution, identifying outliers, understanding data spread, etc.


Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point
is represented as a dot, and the x and y positions of the dots correspond to the values of the two variables.

Common Use Cases: Detecting patterns, correlations, or clusters in data, identifying outliers, exploring 
relationships between two variables, etc.


Heatmap:

Use: Heatmaps are used to represent data in a tabular form where the individual cells are color-coded based
on the values they represent. Heatmaps are particularly useful for visualizing matrices or correlation matrices.

Common Use Cases: Visualizing correlation matrices, representing data with color intensity, displaying 
relationships between variables, etc.


Box Plot (Box-and-Whisker Plot):

Use: Box plots are used to display the distribution of data and to identify outliers. They show the median
, quartiles, and any potential outliers in the data.
Common Use Cases: Comparing distributions of multiple variables, detecting outliers, analyzing data spread
and skewness, etc.
```
